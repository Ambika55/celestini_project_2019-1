```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
imgL = cv2.imread('MiddEval3-data-Q/MiddEval3/trainingQ/Adirondack/im0.png',0)
imgR = cv2.imread('MiddEval3-data-Q/MiddEval3/trainingQ/Adirondack/im1.png',0)

plt.imshow(imgL)
plt.show()
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()
```
